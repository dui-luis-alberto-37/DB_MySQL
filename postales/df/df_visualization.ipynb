```python
import pandas as pd
```

```python
df = pd.read_csv('CPdescarga.txt', header=0, sep='|',encoding='latin1')
df
```

```python
def clean(text):
   if isinstance(text, str):
      to_chage = str.maketrans({
         'Á': 'A',
         'É': 'E',
         'Í': 'I',
         'Ó': 'O',
         'Ú': 'U',
         'á': 'a',
         'é': 'e',
         'í': 'i',
         'ó': 'o',
         'ú': 'u',
         'ñ': 'n',
         'ü': 'u'
      })
      return text.translate(to_chage)
   return text


clean_df = df.map(clean)
clean_df.info()
```

```python
clean_df.to_csv('clean_df.txt', index = False)
```
